```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://movie.douban.com/cinema/nowplaying/shanghai/'
```

```python
fake_header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:61.0) Gecko/20100101 Firefox/61.0'
}
```

```python
html = requests.get(url, headers=fake_header)
```

```python
soup = BeautifulSoup(html.text, 'lxml')
```

```python
nowplay = soup.find('div', id='nowplaying')
upcoming = soup.find('div', id='upcoming')
```

```python
def readnow(div):
    col_name = ['category','title','director','duration','region','actors','score']
    df = pd.DataFrame(columns = col_name)
    now_li = div.find_all('li', {'class':'list-item'})
    for templi in now_li :
        newrow = [templi['data-category'], templi['data-title'], templi['data-director'],
          templi['data-duration'],templi['data-region'],
          templi['data-actors'],templi['data-score']]
        df.loc[now_li.index(templi)] = newrow
    return df
```

```python
def readup(div):
    col_name = ['category','title','director','duration','region','actors','score']
    df = pd.DataFrame(columns = col_name)
    now_li = div.find_all('li', {'class':'list-item'})
    for templi in now_li :
        newrow = [templi['data-category'], templi['data-title'], templi['data-director'],
          templi['data-duration'],templi['data-region'],
          templi['data-actors'],'None']
        df.loc[now_li.index(templi)] = newrow
    return df
```

```python
pd.concat([readnow(nowplay),readup(upcoming)]).reset_index().drop(['index'],axis=1)
```
