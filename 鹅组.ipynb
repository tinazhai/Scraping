```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
url = 'https://www.douban.com/group/blabla/discussion?start=0'
```

```python
def readpost(url):
    fake_header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:61.0) Gecko/20100101 Firefox/61.0'}
    html = requests.get(url, headers=fake_header)
    soup = bs(html.text, 'lxml')
    df = pd.DataFrame(columns=['title','title_link','author','author_link','reply','time'])
    title = soup.find_all('td', class_='title')
    name = soup.find_all('a',{'href': re.compile('https://www.douban.com/people/*')})
    for i in range(0,len(title)):
        df.loc[i] = [title[i].a.get('title'),
                     title[i].a.get('href'),
                    name[i].string,name[i].get('href'),
                    soup.find_all('td', class_='r-count')[i+1].string,
                    soup.find_all('td', class_='time')[i].string]
    return df
```

```python
df = pd.DataFrame()
for i in range(0,5):
    url = 'https://www.douban.com/group/blabla/discussion?start='+str(i*25)
    df = pd.concat([df,readpost(url)]).reset_index().drop(['index'],axis=1)
```
